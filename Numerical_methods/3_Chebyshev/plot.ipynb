```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('plot_data.txt', delimiter=',')
x = data[:300, 1]
y = data[:300, 0]

plt.figure(figsize=(10, 6))
plt.plot(x, y, linestyle='-', color='b')
plt.title('  ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.yscale('log')
plt.grid()
plt.savefig('norm_error_plot_1.png', dpi=300, bbox_inches='tight')
plt.show()
```
